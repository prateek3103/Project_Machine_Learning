**Determine fruit colour score by considering its other attributes**

```python
import pandas as pd
```

```python
fruit=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Fruits.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=fruit['Fruit Colour Score']
```

```python
x=fruit.drop(['Fruit Colour Score','Fruit Name'],axis=1)
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,train_size=0.72,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
